```python
#Import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#read IMDB file into pandas
csv_file2 = "resources/IMDb movies.csv"
IMDB_df = pd.read_csv(csv_file2)
IMDB_df.head()
```

```python
# import movielens movie data
csv_file3 = "resources/movie.csv"
movielens_movie_df = pd.read_csv(csv_file3)
movielens_movie_df.head()
```

```python
# import movielens rating data
csv_file4 = "resources/rating.csv"
movielens_rating_df = pd.read_csv(csv_file4)
movielens_rating_df.head()
```

```python
# get the avg rating for each movieId
movielens_rating_mean = movielens_rating_df.groupby(by='movieId').mean()
movielens_rating_mean.head()
```

```python
# merge movielens df (movie and rating) by column movieId
movielens_df = movielens_movie_df.merge(movielens_rating_mean, how='inner', on="movieId")
movielens_df.head()
```

```python
# drop userId column
movielens_df.drop(columns='userId', inplace=True)
movielens_df.head()
```

```python
#Read rotten tomatoes file into pandas df
csv_file = "resources/rotten_tomatoes.csv"
rottentomatoes_df = pd.read_csv(csv_file)
rottentomatoes_df.head()
```

```python
#Rename title column in order to merge with other dataframes
rottentomatoes_df.rename(columns={'Title':'title'}, 
                 inplace=True)
rottentomatoes_df
```

```python
#Drop duplicate titles in df
rottentomatoes_df.drop_duplicates()
rottentomatoes_df
```

```python
#Sort by ratings in descending order to get the best movies
sorted_df = rottentomatoes_df.sort_values(by='RatingTomatometer', ascending=False)
sorted_df.head()
```

```python
#Only keep movies that have a rating tomatometer score of 70 or greater
good_movies_df = sorted_df.loc[sorted_df['RatingTomatometer'] >= 70]
good_movies_df
```

```python
#Check to see how many movies meet this criteria
goodmovies_df.count()
```

```python
#merging together rotten tomatoes and movielens df
merged_df = rottentomatoes_df.merge(movielens_movie_df, how='inner', on="title")
merged_df
```
